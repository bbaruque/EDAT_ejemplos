# Análisis de Complejidad Algorítmica

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Búsqueda de un dato contenido en el conjunto

En este apartado simplemente cargamos los datos del tiempo tardado en realizar la operación en cada colección
Los datos se han obtenido con la ejecución del programa Java almacenado en este mismo directorio.

Como se puede ver en él, el tiempo medido es el que tarda cada una de las estructuras en conocer si contiene un determinado elemento. Esta operación será siempre análoga en complejidad a la de una búsqueda por elemento en la estructura.

```python
# Carga de los datos
noContenido = pd.read_csv("./TiemposContenido.csv", index_col='tamano',comment='#')
noContenido.head()
```

```python
fig, axarr = plt.subplots(1, 2, figsize=(18, 5))
noContenido.plot.line(ax=axarr[0])
noContenido[['mapa','arbol']].plot.line(ax=axarr[1], ylim=(0,20000))
```

A la izquierda se representa la comparativa entre las 3 colecciones. A la derecha, los mismos datos, pero omitiendo la lista.
